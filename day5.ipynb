```
bottom_line_re = re.compile(r"move (?P<n>\d+) from (?P<src>\d+) to (?P<dst>\d+)")

def main(input):
  input_top, input_bottom = input.split("\n\n")

  top_lines = list(input_top.split("\n"))[::-1]
  stack_numbers, stack_lines = top_lines[0], top_lines[1:]
  stack_ints = list(map(int, stack_numbers.split()))
  stack_data = {stack_number: [] for stack_number in stack_ints}

  for stack_line in stack_lines:
    for c, stack_int in zip(list(stack_line)[1::4], stack_ints):
      if c != " ":
        stack_data[stack_int].append(c)
  
  
  for line in input_bottom.split('\n'):
    mo = bottom_line_re.fullmatch(line)
    assert mo is not None
    n, src, dst = map(int, mo.groups())
    stack_data[dst].extend(stack_data[src][-n:][::-1])
    stack_data[src] = stack_data[src][:-n]
    # print(stack_data)

  return "".join(stack_data[stack_int][-1] for stack_int in stack_ints)
```
